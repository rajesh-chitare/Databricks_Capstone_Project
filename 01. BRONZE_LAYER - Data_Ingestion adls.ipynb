### Transaction Streaming Data Ingestion 

```
# File path in ADLS
file_names = config["paths"]["transactions"]
container_name = config["adls_connection"]["container_name"]
storage_account = config["adls_connection"]["storage_account"]
transaction_path = f"abfss://{container_name}@{storage_account}.dfs.core.windows.net/{file_names}"

def ingest_transaction():
    source_df = (spark.readStream
                        .format("cloudFiles")
                        .option("cloudFiles.format", config["read_stream"]["cloudFiles"]["format"])  
                        .option("header", config["read_stream"]["cloudFiles"]["header"]) 
                        .option("timestampFormat", config["read_stream"]["cloudFiles"]["timestampFormat"])
                        .schema(transactions_schema) 
                        .option("cloudFiles.schemaLocation", config["read_stream"]["cloudFiles"]["schemaLocation"])
                        .option("cloudFiles.inferColumnTypes", config["read_stream"]["cloudFiles"]["inferColumnTypes"])                    
                        .load(transaction_path)
    )

    write_query = (source_df.writeStream
                            .format("delta")
                            .option("checkpointLocation", config["write_stream"]["checkpointLocation"])
                            .option("mergeSchema", config["write_stream"]["mergeSchema"])
                            .outputMode(config["write_stream"]["outputMode"])                         
                            .trigger(processingTime=config["write_stream"]["processingTime"])  
                            .toTable(config["write_stream"]["table"])
    )

    return write_query

ingest_transaction()
```
